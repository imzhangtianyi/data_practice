```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create data
v = 60
t = np.arange(10)
noise = np.random.normal(size=10, scale=5)
y = v * t + noise * t
```

```python
plt.scatter(t, y)
plt.grid()
plt.show()
```

```python
# Train Test split
n = 7
xtr = t[:n]
xte = t[n:]
ytr = y[:n]
yte = y[n:]
```

```python
# Total error
def cal_error(theta, b, x, y):
    se = 0
    N = len(x)
    for i in range(N):
        se += (y[i] - (theta*x[i] + b)) ** 2
    return (se / 2.0) ** .5
```

```python
cal_error(60, 0, xtr, ytr)
```

```python
def gradient(theta, b, x, y, alpha):
    dtheta = 0
    db = 0
    N = len(x)
    for i in range(N):
        dtheta +=  -(2/N) * x[i] * (y[i] - ((theta * x[i]) + b))
        db +=  -(2/N) * (y[i] - ((theta * x[i]) + b))
    return theta - dtheta * alpha, b - db * alpha
```

```python
theta = np.random.randint(100)
b = np.random.randint(100)

plt.figure(figsize=(6,6))
plt.scatter(xtr, ytr, s=60)
plt.plot(t, t*theta + b)
for i in range(n):
    plt.plot([xtr[i], xtr[i]], [xtr[i]*theta + b, ytr[i]], '--', color='r')
plt.xlim(0,10)
plt.ylim(0, 600)
plt.text(1, 500, 'Error={}'.format(np.round(cal_error(theta, b, xtr, ytr),2)), fontsize=16)
plt.show()
```
